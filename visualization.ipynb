```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['Top-1 Accuracy', 'Top-5 Accuracy']
baseline = [76.00, 94.20]
fgsm = [4.40, 31.20]
decrease = [71.60, 63.00]

x = np.arange(len(metrics))  # label locations
width = 0.35  # width of the bars

# Create figure and axis
plt.figure(figsize=(10, 6))
plt.style.use('seaborn')

# Plot bars
rects1 = plt.bar(x - width/2, baseline, width, label='Baseline', color='#2ecc71', edgecolor='black')
rects2 = plt.bar(x + width/2, fgsm, width, label='After FGSM Attack', color='#e74c3c', edgecolor='black')

# Add decrease percentages
for i in range(len(metrics)):
    plt.text(x[i] - width/2, baseline[i] + 2, f'{baseline[i]}%',
             ha='center', va='bottom', fontsize=20, fontweight='bold')
    plt.text(x[i] + width/2, fgsm[i] + 2, f'{fgsm[i]}%',
             ha='center', va='bottom', fontsize=20, fontweight='bold')
    plt.text(x[i] + 0.2, max(baseline[i], fgsm[i]) + 5, f'Decrease: {decrease[i]}%',
             ha='center', va='bottom', fontsize=15, color='#34495e')

# Customize plot
plt.ylabel('Accuracy (%)', fontsize=24, fontweight='bold')
plt.title('Model Performance Before and After FGSM Attack (=0.02)',
          fontsize=24, fontweight='bold', pad=20)
plt.xticks(x, metrics, fontsize=24)
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add legend and adjust layout
plt.legend(loc='upper left', frameon=True, fontsize=20)
plt.tight_layout()

# Add caption
# plt.figtext(0.5, 0.01,
#             'Figure 1: Comparison of model accuracy on clean vs adversarially perturbed test set',
#             ha='center', fontsize=10, style='italic')

plt.show()
```
